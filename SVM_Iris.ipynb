```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head(10)
```

```python
df['target'] = iris.target
```

```python
df['flower name'] = df.target.apply(lambda x : iris.target_names[x])
```

```python
# Creating different dataset for each target category
df0 = df[:50] # sethosa
df1 = df[50:100] # versicolor
df2 = df[100:] # virginica
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='green', marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='blue', marker='.')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], color='yellow', marker='.')
```

```python
# sepal length and sepal width is not good attributes to create svm classifier
# Let's try other features
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],color="green",marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'],color="blue",marker='.')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'],color="orange",marker='.')
```

```python
# As you see petal length and petal width provide us better results

# Let's adjust our dataset to train the SVM classifier

X = df.drop(['target','flower name'], axis='columns')
y = df.target

X
```

```python
y
```

```python
# Let's separate our data into training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict([[4.8,3.0,1.5,0.3]])
```


Support Vector Machines (SVMs) have several important parameters that influence their training and performance. The specific parameters you'll need to tune will depend on the type of SVM you're using (e.g., linear, RBF, etc.) and the nature of your data. However, some key parameters common to most SVMs include:

C: This parameter controls the trade-off between maximizing the margin (separating hyperplane) and minimizing training errors. A higher C prioritizes correctly classifying all training points, potentially leading to a more complex model and overfitting. Conversely, a lower C allows for some margin violations but may result in a simpler, less accurate model.

Kernel: This defines the similarity measure between data points in the feature space. The most common choices are:

Linear kernel: Suitable for linearly separable data.
RBF (Radial Basis Function) kernel: More flexible than the linear kernel, able to handle non-linearly separable data. Requires tuning the gamma parameter, which controls the influence of individual data points.
Polynomial kernel: Can capture complex non-linear relationships but can be computationally expensive for high-dimensional data. Requires tuning the degree parameter (polynomial order).
Other parameters: Depending on the specific SVM implementation, you might encounter additional parameters like:

Epsilon (Îµ) for SVR (Support Vector Regression): Similar to C but controls the tolerance for regression errors.
Shrinking (tol): Controls the convergence tolerance of the optimization algorithm.
Class weights: Assign different weights to different classes for imbalanced datasets.
Tuning these parameters effectively is crucial for optimizing your SVM's performance. Grid search and cross-validation are commonly used techniques to find the optimal parameter combination.
